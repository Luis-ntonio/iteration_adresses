```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/out/ad1.csv', index_col=0)
df2 = pd.read_csv('data/out/ad2.csv', index_col=0)
df3 = pd.read_csv('data/out/ad3.csv', index_col=0)
df4 = pd.read_csv('data/out/ad4.csv', index_col=0)
df5 = pd.read_csv('data/out/ad5.csv', index_col=0)
df6 = pd.read_csv('data/out/ad6.csv', index_col=0)
```

```python
df2 = df2.drop(columns=['X_x', 'Y_x', 'NUEVA DIRECCION_x', 'COMUNA DOMICILIO_x', 'DIRECCION TOTAL_x'])
df2 = df2.rename(columns={'X_y': 'X', 'Y_y': 'Y', 'NUEVA DIRECCION_y': 'NUEVA DIRECCION', 'COMUNA DOMICILIO_y': 'COMUNA DOMICILIO', 'DIRECCION TOTAL_y': 'DIRECCION TOTAL'})
df2
```

```python
res = pd.concat([df1, df2, df3, df4, df5, df6])
res = res.sort_index()
```

```python
res
```

```python
df_f = pd.read_csv('./data/direcciones.csv', sep=';')
```

```python

# Merge df_f with res on 'index'
df_f = df_f.merge(res, on='DIRECCION DE DOMICILIO', how='left')
```

```python
df_f.to_excel('data/out/merged.xlsx', index=False)
```
