```python
from validate_adrresses import validate_address, remove_duplicates_preserve_order
import pandas as pd
import os
from tqdm import tqdm
print(os.getcwd())
```

```python
df = pd.read_csv('./data/direcciones-consolidado.csv', sep=';', encoding='latin1')
```

```python
cols = [col for col in df]
col1 = cols[:3]
col2 = cols[3:]

cols = []
for col in col1:
    cols.append(col)
cols.append("NUEVA DIRECCION")
cols.append("FLG_ITER")
for col in col2:
    cols.append(col)
```

```python
df["NUEVA DIRECCION"] = None
df = df.reindex(columns=cols)
df = df[df["IND TERRITORIOS"] == '1']
df
```

```python
df_tmp = df[["NOMBRE", "DIRECCION DE DOMICILIO", "COMUNA DOMICILIO", "NUEVA DIRECCION", "X", "Y", "FLG_ITER"]].copy()
df_tmp["DIRECCION TOTAL"] =  df_tmp.apply(lambda row: remove_duplicates_preserve_order(row["DIRECCION DE DOMICILIO"] + " " + row["COMUNA DOMICILIO"], row["COMUNA DOMICILIO"]),axis=1)
df_tmp
```

```python
arr_df = []
itr = 5
step = len(df_tmp)//itr
for i in range(itr):
    if i < itr - 1:
        arr_df.append(df_tmp[step*i:step*(i+1)])
    else:
        arr_df.append(df_tmp[step*i:])
arr_df[-1]
```

```python
import dask.dataframe as dd
import pandas as pd
import numpy as np
from multiprocessing import Pool

def parallel(df_):
    ddata = dd.from_pandas(df_, npartitions=30)

    # Sample meta dataframe to indicate the structure of the output
    meta = pd.DataFrame({
        "NOMBRE": pd.Series(dtype='str'),
        "DIRECCION DE DOMICILIO": pd.Series(dtype='str'),
        "COMUNA DOMICILIO": pd.Series(dtype='str'),
        "NUEVA DIRECCION": pd.Series(dtype='str'),
        "X": pd.Series(dtype='float'),
        "Y": pd.Series(dtype='float'),"FLG_ITER": pd.Series(dtype='str'),
        "DIRECCION TOTAL": pd.Series(dtype='str')
        
    })

    def apply_validate_address(df_chunk):
        # Handle missing values before passing to validate_address
        df_chunk[["DIRECCION TOTAL"]] = df_chunk[["DIRECCION TOTAL"]].fillna("")
        
        # Apply the validation function to each row and unpack the results
        df_chunk[["NUEVA DIRECCION", "X", "Y", "FLG_ITER"]] = df_chunk.apply(
            lambda row: pd.Series(validate_address(row["DIRECCION TOTAL"], row["COMUNA DOMICILIO"])),
            axis=1
        )
        return df_chunk

    # Map the function to each partition, with the meta keyword to define the structure
    res = ddata.map_partitions(apply_validate_address, meta=meta).compute()
    return res

def parallel_wrapper(df):
    return parallel(df)

if __name__ == '__main__':
    arr_df_tmp = []
    arr_df_tmp.append(parallel_wrapper(df_tmp))
```
